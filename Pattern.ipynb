```python
def hollow_diamond(n):
    # Upper part
    for i in range(n):
        for j in range(n - i - 1):
            print(" ", end="")
        for j in range(2 * i + 1):
            if j == 0 or j == 2 * i:
                print("*", end="")
            else:
                print(" ", end="")
        print()
    
    # Lower part
    for i in range(n - 2, -1, -1):
        for j in range(n - i - 1):
            print(" ", end="")
        for j in range(2 * i + 1):
            if j == 0 or j == 2 * i:
                print("*", end="")
            else:
                print(" ", end="")
        print()

# Example usage
hollow_diamond(5)
```

```python
def number_pyramid(n):
    for i in range(1, n + 1):
        print(" " * (n - i), end="")
        for j in range(1, i + 1):
            print(j, end=" ")
        print()

number_pyramid(5)
```

```python
def inverted_number_pyramid(n):
    for i in range(n, 0, -1):
        print(" " * (n - i), end="")
        for j in range(1, i + 1):
            print(j, end=" ")
        print()

inverted_number_pyramid(5)
```

```python
def floyd_triangle(n):
    num = 1
    for i in range(1, n + 1):
        for j in range(i):
            print(num, end=" ")
            num += 1
        print()

floyd_triangle(5)
```

```python
def diamond_numbers(n):
    # Upper part
    for i in range(1, n + 1):
        print(" " * (n - i), end="")
        for j in range(1, i + 1):
            print(j, end=" ")
        print()
    # Lower part
    for i in range(n - 1, 0, -1):
        print(" " * (n - i), end="")
        for j in range(1, i + 1):
            print(j, end=" ")
        print()

diamond_numbers(5)
```

```python
def pascal_triangle(n):
    for i in range(n):
        print(" " * (n - i), end="")
        val = 1
        for j in range(i + 1):
            print(val, end=" ")
            val = val * (i - j) // (j + 1)
        print()

pascal_triangle(5)
```

```python
def username_rank(username: str) -> str:
    upper_count = 0
    digit_count = 0
    special_count = 0
    
    for ch in username:
        if ch.isupper():
            upper_count += 1
        elif ch.isdigit():
            digit_count += 1
        elif not ch.isalnum():
            special_count += 1
            
    return f"UPPER:{upper_count} DIGITS:{digit_count} SPECIAL:{special_count}"
if __name__ == "__main__":
    try:
        username = input("Enter the username: ").strip()
        if not username:
            raise ValueError("Input cannot be empty")
        
        result = username_rank(username)
        print(result)
    
    except Exception as e:
        print("Error:", e)
```

```python
def digital_root(n):
    if n < 0:
        return "Negative numbers not allowed"
    
    while n > 9:
        total = 0
        for digit in str(n):
            total += int(digit)
        n = total
    
    return n

n = int(input())
print(digital_root(n))
```

```python
# Parse input: first line has n, then n pairs of (flavor, qty), then flavor to update, then new qty
data = input().split()
n = int(data[0])

# store as list of [name, qty]
ice = []
idx = 1
for _ in range(n):
    name = data[idx]
    qty = int(data[idx + 1])
    ice.append([name, qty])
    idx += 2

flavor = data[idx]
new_stock = int(data[idx + 1])

# update the matching flavor
for item in ice:
    if item[0] == flavor:
        item[1] = new_stock

# 1) print converted list (after modification)
print(ice)

# 2) convert to tuple of tuples and print
ice_tuple = tuple((name, qty) for name, qty in ice)
print(ice_tuple)
```

```python
s = input()

def longest(s):
    ans = s[0]
    for i in range(len(s)):
        for j in range(i, len(s)):
            sub = s[i:j+1]
            if sub == sub[::-1] and len(sub) > len(ans):
                ans = sub
    return ans

print(longest(s))
```
