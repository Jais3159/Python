```python
# If statements and comparisons:
def max_num(a,b,c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c
    
print(max_num(3,4,5))
```

```python
# Modified
secret_word = "giraffe"
guess = ""
guess_count = 0
guess_limit = 3
out_of_guesses = False

while guess != secret_word and not(out_of_guesses):
    if guess_count < guess_limit:
        guess = input("Enter guess: ")
        guess_count += 1
    else:
        out_of_guesses = True

if out_of_guesses:
    print("Out of Guesses, YOU LOSE")
else:
    print("You win!")
```

```python
# Exponent function
def raise_to_power(base_num, pow_num):
    result = 1
    for index in range(pow_num):
        result *= base_num
    return result
print(raise_to_power(2, 3))
```

```python
# Build a translator
# Giraffe Language
# vowels -> g
# ------------

# dog -> dgg
# cat -> cgt
def translate(phrase):
    translation = ""
    for letter in phrase:
        if letter.lower() in "aeiou":
            if letter.upper():
                translation += "G"
            else:
                translation += "g"
        else:
            translation += letter
    return translation

print(translate(input("Enter a phrase: ")))
```

```python
# Try Except
try:
    number1 = int(input("Enter the number: "))
    number2 = int(input("Enter the number: "))
    value = number1 / number2
    print(value)
except ValueError:
    print("Error: Enter the valid numeric value.")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
finally:
    print("Attempt completed")
```
