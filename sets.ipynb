```python
languages = {'C','C++','Java','Python'}
print('Original Set: ', languages)
Removedvalue = languages.remove('Java') # we can also use discard instead of remove
print('Updated Set: ', languages)
```

```python
languages = {'C','C++','Java','Python'}
print('Orginal set:', languages)
removedvalue = languages.remove('.NET')
print('Updated Set: ', languages)
```

```python
languages = {'C','C++','Java','Python'}
print('Original set:',languages)
removedvalue = languages.pop()
print(removedvalue)
print('Updated set:', languages)
```

```python
s1 = {'Apple','Banana','Mango'}
s2 = {'Google''Microsoft','Apple'}
s1.intersection_update(s2)
print(s1)
```

```python
s1 = {'Apple','Banana','Mango'}
s2 = {'Google''Microsoft','Apple'}
s3 = s1.difference(s2)
print(s3)
```

```python
s1 = {'Apple','Banana','Mango'}
s2 = {'Google','Microsoft','Apple'}
s3 = s1.symmetric_difference(s2)
print(s3)
```
