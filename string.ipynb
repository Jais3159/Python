```python
# Anagram pairs in the string / Bubble sorting
words = ['cat', 'dog', 'tac', 'god', 'act']
for i in range(len(words)):
    for j in range(i + 1, len(words)):
        list1 = [ch for ch in words[i]]
        list2 = [ch for ch in words[j]]
        for x in range(len(list1)):
            for y in range(x + 1, len(list1)):
                if list1[x] > list1[y]:
                    list1[x], list1[y] = list1[y], list1[x]
        
        for x in range(len(list2)):
            for y in range(x + 1, len(list2)):
                if list2[x] > list2[y]:
                    list2[x], list2[y] = list2[y], list2[x]
        
        if list1 == list2:
            print(words[i], '<->', words[j])
```

```python
# A cyber security have to validate the password the condition is password must contain atleast 1 Uppercase
# and 1 lowercase letter.
# test cases:           output
# Input:'Python'        strong
# Input:'hello'         weak
# Input:'Armaan123'     strong
password = input("Enter your password: ")
has_upper = any(char.isupper() for char in password)
has_lower = any(char.islower() for char in password)
    
if has_upper and has_lower:
    print("strong")
else:
    print("weak")
```
