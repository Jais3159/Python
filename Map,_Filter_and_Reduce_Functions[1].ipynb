```python
#USING MAP FUNCTION (APPLIES THE FUNTION TO ALL ELEMENTS OF THE LIST)

def double(x):
    return x*2

nums=[1,2,3,4,5]
doubled_nums=list(map(double,nums))
print(doubled_nums)
```

```python
#USING MAP FUNCTION TO MULTIPLY NUMBERS WITH A SPECIFIC NUMBER

def multnum(nums):
    return nums*3   
nums=map(multnum,[1,2,3,4,5])
print("Multiplying each number by 3:",list(nums))
```

```python
#MAPS APPLICATION USING LAMBDA

numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, numbers))
print(squares)  
```

```python
#REDUCE FUNTIONS:The reduce func is a high oreder func and it is applied to the items of the itrable reducing the iterable for a single value it is a part of func tools module which is aldredy in pyhton and it needs to be imported before using

from functools import reduce
nums=[1,2,3,4,5]
res = reduce(lambda x, y: x + y,nums)
print(res)
```

```python
#Find Max and Min element using reduce

from functools import reduce
nums=[3,7,9,11,15,4,2,1,8]
res_min=reduce(lambda x,y: x if x<y else y,nums)
res_max=reduce(lambda x,y: x if x>y else y,nums)

print("Min:",res_min)
print("Max:",res_max)
```
