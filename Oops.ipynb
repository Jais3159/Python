```python
class BankAccount:
    def __init__(self,initial_balance):
        self._balance = initial_balance
    def get_balance(self):
        return self._balance
    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            return f"Deposited: {amount}. New Balance: {self._balance}"
        else:
            return "Amount must be positive."
account = BankAccount(100)
print(account.deposit(50))
try:
    print(account._balance)
except AttributeError as e:
    print(f"Error: {e}")
```

```python
class Employee:
    name = ""
    age = 0
    salary = 0
    EPFO = 0
govt = Employee()
private = Employee()
govt.name = "Rashmeet"
govt.age = 18
govt.salary = 100000
govt.EPFO = 3200
private.name = "Harsh"
private.age = 60
private.salary = 80000
private.EPFO = 70000
print(f"Name : {govt.name} Age : {govt.age} Salary : {govt.salary} EPFO : {govt.EPFO}")
print(f"Name : {private.name} Age : {private.age} Salary : {private.salary} EPFO : {private.EPFO}")
```
