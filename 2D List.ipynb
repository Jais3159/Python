```python
#2
grid = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]

total = sum(sum(row) for row in grid)
print('sum of all elements: ', total)
count = sum(len(row) for row in grid)
average = total / count
print('Average of all elements: ', average)

minimum_value = grid[0][0]
maximum_value = grid[0][0]
for row in grid:  # row ke liye
    for value in row:   # column ke liye
        if value < minimum_value:
            minimum_value = value
        if value > maximum_value:
            maximum_value = value
print('Minimum element: ', minimum_value)
print('Maximum element: ', maximum_value)
```

```python
#5 FOUNDING NUMBER
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
x = 7
found = False   # boolean expression

for i in range(len(matrix)):    # rows ke liye
    for j in range(len(matrix[i])): # columns ke liye
        if matrix[i][j] == x:
            print(f'Element {x} found at row {i}, column {j}')
            found = True
            break
    if found:
        break
if not found:
    print(f'Element {x} not found in the following matrix')
```

```python
#6
matrix = [
    [1,2,3],
    [4,5,5],
    [7,5,9]
]
x = 5
positions = []

for i in range(len(matrix)):
    for j in range(len(matrix[i])):
        if matrix[i][j] == x:
            positions.append((i,j))

if positions:
    print(f'Element {x} found at positions: {positions}')
else:
    print(f'Element {x} not found')
```
